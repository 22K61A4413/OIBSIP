```python
# Pie Chart - Unemployment distribution by region
plt.figure(figsize=(8, 8))
unemployment_data['Region'].value_counts().plot.pie(autopct='%1.1f%%', cmap='Set3')
plt.title("Unemployment Distribution by Region")
plt.show()
```

```python
# Bar Graph - Unemployment rate per state
plt.figure(figsize=(12, 6))
# Replace 'Unemployment_rate' with the actual column name from your DataFrame.
# For example if your column name is ' Estimated Unemployment Rate (%)', then:
sns.barplot(data=unemployment_data, x='Region', y=' Estimated Unemployment Rate (%)', ci=None)
plt.xticks(rotation=90)
plt.title("Unemployment Rate by Region")
plt.xlabel("Region")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Histogram - Distribution of unemployment rates
plt.figure(figsize=(10, 5))
# Use the correct column name ' Estimated Unemployment Rate (%)'
sns.histplot(unemployment_data[' Estimated Unemployment Rate (%)'], bins=20, kde=True, color='blue')
plt.title("Distribution of Unemployment Rates")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box Plot - Compare unemployment rates in Urban vs Rural areas
plt.figure(figsize=(8, 5))
# Use the correct column name ' Estimated Unemployment Rate (%)' for the 'y' axis
sns.boxplot(data=unemployment_data, x='Area', y=' Estimated Unemployment Rate (%)', palette='Set2')
plt.title("Unemployment Rate Comparison: Urban vs Rural")
plt.xlabel("Area")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
# Heatmap - Correlation between numerical features
plt.figure(figsize=(8, 6))
# Select only numerical features for correlation calculation
numerical_data = unemployment_data.select_dtypes(include=np.number)
sns.heatmap(numerical_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
